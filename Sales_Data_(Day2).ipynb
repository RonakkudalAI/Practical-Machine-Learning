```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("/content/sales_data.csv")
df
```

```python
df['Category'].value_counts()
```

```python
#Data Cleaning
df.isnull().sum()
```

```python
df.fillna(0,inplace = True)
```

```python
#Grouping the data
category_sales=df.groupby("Category")['Revenue'].sum()
```

```python
#Identify the revnue by Category
monthly_sales  = df.groupby("Month")["Revenue"].sum()
```

```python
#Average Order value

ave_order = df["Revenue"].mean()
```

```python
#Max, Min and Average
revenue_array = df["Revenue"].values
np.max((revenue_array))
np.min((revenue_array))
np.mean((revenue_array))
np.median((revenue_array))
```

```python
# Bar Graph
category_sales.plot(kind="bar")
plt.title("Category wise Revenue")
plt.xlabel("Category")
plt.ylabel("Revenue")
plt.show()
```

```python

monthly_sales .plot(kind="line", marker='o')
plt.title("Monthly Sales Trends")
plt.xlabel("Months")
plt.ylabel("Revenue")
plt.show()
```

```python
#Distribution Plot
sns.histplot(df["Revenue"],kde = True)
plt.title("Distribution plot of Revenue")
plt.show()
```

```python
#Box Plot
sns.boxplot(x="Category",y = "Revenue",data = df)
plt.title("Revenue By Category")
```

```python
sns.scatterplot(x = "Price",y ="Revenue" ,hue="Category", data = df)
```

```python
sns.scatterplot(x = "Category",y ="Revenue" ,hue="Category", data = df)
```

```python
sns.scatterplot(x = "Quantity",y ="Revenue" ,hue="Category", data = df)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

corr = df.corr(numeric_only=True)

sns.heatmap(corr, annot=True, cmap="coolwarm")

plt.title("Correlation Heatmap")
plt.show()
```
