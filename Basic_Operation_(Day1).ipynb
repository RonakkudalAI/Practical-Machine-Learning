Dictionary : Key-Value , heterogenous , no order , mutable and duplicate key not not allowed


```python
customer = {
    "Name":["Ronak","Shivam","Rahul"],
    "City" :["Udaipur","Mp","Jaipu"],
    "Age" : [20,22,24]
}
customer
```

```python
customer['Name']
```

```python
customer.get('Age')
```

```python
for key,value in customer.items():
  print(key,value)
```

```python
del customer["Age"]
```

```python
customer
```

Sets : Unordered , Unique, not mutable

```python
cusines = {"Indian","Chinese","Italian","Indian"}
cusines
```

```python
cusines.add("Mexican")
cusines
```

```python
cusines.discard("Chinese")
cusines
```

```python
cusines_set1  = {"Indian","Chinese","Italian"}
cusines_set2 = {"Mexican","Chinese","Italian"}
```

```python
cusines_set1.union(cusines_set2)
```

```python
cusines_set1.intersection(cusines_set2)
```
